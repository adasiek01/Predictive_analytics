```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Sample - Superstore.xls')

df_orders = pd.read_excel('Sample - Superstore.xls', sheet_name='Orders')
df_returns = pd.read_excel('Sample - Superstore.xls', sheet_name='Returns')
df_people = pd.read_excel('Sample - Superstore.xls', sheet_name='People')
```

```python
Profit = df_orders["Profit"]
Order_dates = pd.to_datetime(df_orders['Order Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Order_dates, Profit)
plt.title('Profit over days')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
